# Recommendation System (Book)

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
# Import Dataset
bk = pd.read_csv('book.csv', encoding='latin-1')
bk
```

```python
bk1 = bk.drop(['Unnamed: 0'], axis=1).rename(columns={'User.ID':'User_ID', 'Book.Title':'Book_Title', 'Book.Rating':'Book_Rating'})
```

```python
bk1
```

```python
# Preprocessing
```

```python
# no. of unique users in the data
User_ID_unique = bk1.User_ID.unique()
```

```python
User_ID_unique = pd.DataFrame(User_ID_unique)
```

```python
len(bk1.Book_Title.unique())
```

```python
user_bk = bk1.pivot_table(index='User_ID',
                         columns='Book_Title',
                         values='Book_Rating')
```

```python
user_bk
```

```python
# Build The Model
```

```python
user_bk.index = bk1.User_ID.unique()
```

```python
# Impute Those NaNs with 0 values
```

```python
user_bk.fillna(0, inplace=True)
```

```python
user_bk
```

```python
# Calculate cosine similarity between users
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_similar = 1 - pairwise_distances(user_bk.values,metric='cosine')
```

```python
user_similar
```

```python
user_similar.shape
```

```python
# Store the results in a dataframe
```

```python
user_similar_df = pd.DataFrame(user_similar)
```

```python
# Set the index and column names to user ids
user_similar_df.index = bk1.User_ID.unique()
user_similar_df.columns = bk1.User_ID.unique()
```

```python
user_similar_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_similar, 0)
user_similar_df.iloc[0:5, 0:5]
```
