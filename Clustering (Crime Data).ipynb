# Clustering (Crime Data)

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# Import Dataset
crime = pd.read_csv('crime_data.csv')
crime
```

```python
# Normalized data function
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
df_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
```

```python
y_hc
```

```python
crime['h_clusterid']=hc.labels_
crime
```

```python
# KMeans
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crime1=pd.read_csv('crime_data.csv')
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
# Selecting 4 cluster from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Clust']=x
```

```python
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.Clust).mean()
```

```python
# DBSCAN
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime3=pd.read_csv('crime_data.csv')
```

```python
crime3.info()
```

```python
df=crime3.iloc[:,1:5]
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
```

```python
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([crime3,cl],axis=1)
```
